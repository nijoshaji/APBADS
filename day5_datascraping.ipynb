```python
import csv
with open ("sales_data.csv","r") as file:
    #print(file)
    reader = csv.DictReader(file)
    salesdata = (list(reader))
    with open ("product_data.csv","r") as file:
        reader = csv.DictReader(file)
        productdata = (list(reader))
        
        ### print both csv files
        #print(salesdata)
        #print(productdata)
#### calculating total for each sales transaction(each row)
###### find product name from 2nd file for each sale using product id as the key (using function)
        
        def productname(id):
            for product in productdata:
                if product["Product_ID"] == id:
                    return product["Product_Name"] 

        for sale in salesdata:
            row_total = int(sale["Quantity"]) * float(sale["Price_per_Unit"])
            print(productname(sale["Product_ID"]), sale["Product_ID"], sale["Quantity"], row_total)
```

```python
##### map(lambda function, list) - takes 2 arguments and iterates the lambda through all elements of the list

numbers = [2,3,4,5,6]
double = list(map(lambda x: x*2, numbers))
print(double)

square = list(map(lambda x: x*x, numbers))
print(square)
```
